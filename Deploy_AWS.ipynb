## Baixando e configurando ambiente para o aplicativo Streamlit
image.png

Depois de realizar a conexão é necessário atualizar o ambiente, para isso é preciso entrar no modo adimnistrador utilizando ```sudo su``` e ```yum update```.

### Git e Github
Agora é preciso instalar o git com o comando ```yum install git```
Instalar o pip ```yum install python3-pip```.

Utilizar o comando ```git clone url-do-repositório```. Caso dê errado o comando uma opção é utilizar a chave SSH, mas e ainda sim o erro persistir é possível criar uma SSH no console da AWS atual e assim passar para o github.

Para criar a chave SSH:

```ssh-keygen -t ed25519 -C "your_email@example.com"```

Usar o comando para recuperar a chave, após isso ser feito é necessário copiar a chave e adicionar ao Github:

```cat ~/.ssh/id_ed25519.pub```

### Instalando as dependências
Instalar as dependências do APP ```python3 -m pip install -r requirements.txt``` e o Streamlit usando o seguinte comando para burlar o erro do console ```python3 -m pip install --ignore-installed streamlit```.

Agora é rodar com o comando ```python3 -m streamlit run app.py```.
